Simple Linear Regression: y = ax + b

# Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

# Data Generation

```python
rng = np.random.RandomState(19)
x = 10*rng.rand(50)
y = 2*x-5 + rng.randn(50)
plt.scatter(x,y)
```

# Linear Regression Estimator

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True) # fit_intercept=True => calculate intercept too

x_2D = x[:, np.newaxis] # LinearRegression needs x in 2D shape
model.fit(x_2D, y)
```

```python
xfit = np.linspace(start=0, stop=10, num=1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit, color='blue')
```

```python
print(f'Model\'s Slope: {round(model.coef_[0],2)}; Model\'s Intercept: {round(model.intercept_,2)}')
```
