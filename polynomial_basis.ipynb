Basis Function: Transforms data so that linear regression can be applied to non-linearly related variables

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

## Data Generation

```python
rng = np.random.RandomState(19)

x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)
```

## Pipeline

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

poly_model_5 = make_pipeline(PolynomialFeatures(5), LinearRegression())
poly_model_7 = make_pipeline(PolynomialFeatures(7), LinearRegression())
poly_model_19 = make_pipeline(PolynomialFeatures(19), LinearRegression())
```

# Regression

```python
poly_model_7.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = poly_model_7.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit)
```

```python
poly_model_5.fit(x[:, np.newaxis], y)

yfit = poly_model_5.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit)
```

```python
poly_model_19.fit(x[:, np.newaxis], y)

yfit = poly_model_19.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit)
```
