Basis Function: Transforms data so that linear regression can be applied to non-linearly related variables

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

## Data Generation

```python
rng = np.random.RandomState(19)

x = 10 * rng.rand(50)
y = np.sin(x) + 0.1 * rng.randn(50)
```

## Pipeline

```python
from utils import GaussianFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

gauss_model_15 = make_pipeline(GaussianFeatures(15), LinearRegression())
gauss_model_20 = make_pipeline(GaussianFeatures(20), LinearRegression())
gauss_model_30 = make_pipeline(GaussianFeatures(30), LinearRegression())
```

# Regression

```python
gauss_model_15.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = gauss_model_15.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit)
```

```python
gauss_model_20.fit(x[:, np.newaxis], y)

yfit = gauss_model_20.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit)
```

```python
gauss_model_30.fit(x[:, np.newaxis], y)

yfit = gauss_model_30.predict(xfit[:, np.newaxis])

plt.scatter(x,y)
plt.plot(xfit, yfit)
```
